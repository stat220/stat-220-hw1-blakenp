**Problem 2**. Analyze the Carbon_West data file using the following code.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression
carbon_west = pd.read_fwf("https://stat4ds.rwth-aachen.de/data/Carbon_West.dat")
carbon_west.head()
```

Part a. Construct a histogram of CO2 emissions. 

```python
plt.hist(carbon_west)
plt.xticks(rotation=90)
plt.title("Carbon West Data")
plt.xlabel("Nation")
plt.ylabel("CO2")
plt.show()
```

Part b. Find the mean, median, and standard deviation. 

```python
print(f"Mean = {np.mean(carbon_west["CO2"])}")
print(f"Median = {np.median(carbon_west["CO2"])}")
print(f"Standard Deviation = {np.std(carbon_west["CO2"])}")
```

**Problem 3**. Read in the murder data set using the following code. The variable murder is the murder rate (per 100,000 population) for each state in the U.S. in 2017 according to the FBI Uniform Crime Reports. At first, do not use the observation for D.C. (DC). 

```python
murders = pd.read_fwf("https://stat4ds.rwth-aachen.de/data/Murder.dat")
murders.head()
```

Part a. Find the mean and standard deviation for murders without D.C.

```python
filtered = murders[murders["state"] != "DC"]

print(f"Mean without DC = {np.mean(filtered["murder"])}")
print(f"Standard Deviation without DC = {np.std(filtered["murder"])}")
```

Part b. Find the five-number  summary without D.C. 

```python
summary = filtered["murder"].describe()

print(f"Five-number Summary without DC:")
print(f"min: {summary["min"]}")
print(f"Q1: {summary["25%"]}")
print(f"median: {summary["50%"]}")
print(f"Q3: {summary["75%"]}")
print(f"max: {summary["max"]}")

print("\ndescribe() summary with more values:")
print(summary)
```

Part c. Now include the observation for D.C. and recalculate the mean, median, and five-number summary. What is affected more by this outlier: The mean or the median? The range or the inter-quartile range?

```python
new_summary = murders["murder"].describe()

print("Mean, Median, and Five-number Summary with DC:")
print(f"mean: {new_summary["mean"]}")
print(f"median: {new_summary["50%"]}")
print("\nFive-number Summary:")
print(f"min: {new_summary["min"]}")
print(f"Q1: {new_summary["25%"]}")
print(f"median: {new_summary["50%"]}")
print(f"Q3: {new_summary["75%"]}")
print(f"max: {new_summary["max"]}")
```

## Effects on Mean and Median
The **Mean** is affected more than the **Median** in this case since prior to adding the *DC* outlier, the **mean** was just **4.74** and then jumped to about **5.25**. The **median**, on the other hand, simply changed from **4.85** to **5.0**.
## Effects on the range and inter-quartile range
The **range** had much more drastic change with the addition of *DC* data than the **inter-quartile range**. The original **range** was simply **11.4** but became **23.2** after the *DC* addition, while the **IQR** changed from **3.55** to **3.8**, which isn't a big change.

**Problem 4**. Read in the Houses data file using the following code.  The variables include price (thousands of dollars), size (square feet), tax bill (dollars), number of bathrooms, number of bedrooms, and whether the house is new (1 = yes, 0 = no) for 100 home sales in Gainesville, Florida. Let
y = selling price and x = size of home.

```python
houses = pd.read_fwf("https://stat4ds.rwth-aachen.de/data/Houses.dat")
houses.head()
```

Part a. Construct a scatterplot between square feet and price of a home. 

```python
x = houses[["size"]]
y = houses["price"]

plt.scatter(x, y)
plt.title("House Size and Price Plot")
plt.xlabel("Size")
plt.ylabel("Price")
plt.show()
```

Part b. Find the correlation between the two variables.

```python
print(f"Correlation between House Size and House Price = {houses["size"].corr(houses["price"])}")
```

Part c. Find and plot the regression line on top of the scatterplot. 

```python
model = LinearRegression()
model.fit(x, y)
plt.scatter(x, y, alpha=0.7, marker="*", label="Scatterplot")

# Plot the regression line
plt.plot(x, model.predict(x), color="red", label="Regression Line")

plt.title("Scatter Plot with Regression Line")
plt.xlabel("Size")
plt.ylabel("Price")
plt.legend()
plt.grid(True)
plt.show()
```

Part d. Create a side by side boxplot of price comparing homes that are new and homes that are not new. 

```python
houses.boxplot(column="price", by="new", vert=False,grid = False)
plt.title("Side-by-Side Box Plot of Price by Home Age")
plt.xlabel("Price")
plt.ylabel("Home Age")
plt.show()
```

Part e. What is the relationship between size and price.  Justify your answer using the plots created. Do the same for the relationship between price and whether or not a house is new, again using the plot to justify your answer.

### Relationship Between Size and Price
Based on the graphs, it seems like size and price have a positive linear relationship. That is to say, as house size increases, so does the price of the house. Also, the correlation value of **.83** between size and price is very high, further showing that these values are correlated and have a positive relationship. The regression line highlights this as well, and even though there are not as many big houses as there are small ones, the bigger the house, the higher the price is.
### Relationship Between Price and Whether or Not a House is New
According to the side-by-side box plot above, it appears that on average newer houses are more expensive than older houses. This is because the box plot of newer houses has greater values for all five of the *five-number summary* values. The median price for newer houses is much higher than the median price for older houses. There are, however, some outliers among the older houses that are still expensive, and one of these outliers is actually more expensive than the most expensive house from the new houses data, which is interesting. Overall though, it seems that the status of a house being *new* is much more likely on average to indicate higher prices than houses that are not new.
